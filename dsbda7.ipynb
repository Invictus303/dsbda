```python
import nltk
nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import TfidfVectorizer

text = "Data analytics involves collecting, processing, and analyzing large amounts of data to discover patterns, trends, and insights that help organizations make informed decisions."

print("\nOriginal Text:\n", text)
```

```python
tokens = word_tokenize(text)
print("\nTokens:\n", tokens)
```

```python
pos_tags = pos_tag(tokens)
print("\nPOS Tags:\n", pos_tags)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("\nAfter Stop Words Removal:\n", filtered_tokens)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Words:\n", lemmatized_words)

```

```python
documents = [
    text,
    "Machine learning helps in data analysis",
    "Natural language processing is part of AI"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("\nTF-IDF Matrix:\n")
print(tfidf_matrix.toarray())

print("\nFeature Names:\n", vectorizer.get_feature_names_out())
```
