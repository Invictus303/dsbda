```python
data = """Temperature,Dew Point,Wind Speed
72.0,60.0,10.0
68.5,55.0,15.5
75.3,63.0,8.0
70.1,58.5,12.0
65.8,52.0,20.0
64.2,51.0,7.0
76.1,64.0,18.0
69.0,56.0,11.0
71.2,59.0,14.0
70.0,57.5,13.5
"""

with open("sample_weather.txt", "w") as f:
    f.write(data)

print("File created successfully!")
```

```python
# ==============================
# 1. Import Libraries
# ==============================
import pandas as pd
import numpy as np

# ==============================
# 2. Load Dataset
# ==============================
df = pd.read_csv('sample_weather.txt', delimiter=',')

print("First 5 rows of dataset:")
display(df.head())
```

```python

# ==============================
# 3. Feature Types
# ==============================
print("\nFeature Types:")
print(df.dtypes)

print("\nFeature Description:")
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"{col} → Categorical (Nominal)")
    else:
        print(f"{col} → Numeric (Continuous)")
```

```python
# ==============================
# 4. Calculate Averages
# ==============================
print("\nCalculating Averages...")

avg_temp = df['Temperature'].mean()
avg_dew = df['Dew Point'].mean()
avg_wind = df['Wind Speed'].mean()

print(f"Average Temperature : {avg_temp:.2f}")
print(f"Average Dew Point : {avg_dew:.2f}")
print(f"Average Wind Speed : {avg_wind:.2f}")
```

```python
# ==============================
# 5. Statistical Summary
# ==============================
print("\nStatistical Summary:")
print(df.describe())
```

```python

# ==============================
# 6. Final Inference
# ==============================
print("\nFinal Inference:")
print("1. Dataset successfully read from sample_weather.txt.")
print(f"2. Average Temperature = {avg_temp:.2f}")
print(f"3. Average Dew Point = {avg_dew:.2f}")
print(f"4. Average Wind Speed = {avg_wind:.2f}")
print("5. All features are Numeric (Continuous) in nature.")
```
