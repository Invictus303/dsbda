```python
# ==============================
# 1. Import Libraries
# ==============================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================
# 2. Load Titanic Dataset
# ==============================
df = sns.load_dataset('titanic')

print("First 5 rows:")
display(df.head())
```

```python
# ==============================
# 3. Check Basic Info
# ==============================
print("\nDataset Info:")
print(df.info())
```

```python

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ==============================
# 4. Boxplot: Age vs Gender + Survival
# ==============================
plt.figure(figsize=(8,6))

sns.boxplot(
    x='sex',
    y='age',
    hue='survived',
    data=df
)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```
